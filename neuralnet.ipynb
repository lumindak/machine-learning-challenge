```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
exo = pd.read_csv('exoplanet_cleaned.csv')
exo
```

```python
#X and y values
#drop y vales and string object from df
X = exo.drop(columns=['koi_disposition']).values
y = exo["koi_disposition"].values

print(X.shape, y.shape)
```

```python
#separate 'training' and 'testing' data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
y_train
```

```python
# data_train = y_train.copy()
# data_test = y_test.copy()

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
encoded_y_test
```

```python
from keras.utils import to_categorical

#One-hot encoding
one_hot_y_train = to_categorical(encoded_y_train)
one_hot_y_test = to_categorical(encoded_y_test)
one_hot_y_train
```

```python
from sklearn.preprocessing import MinMaxScaler
X_scale = MinMaxScaler().fit(X_train)

X_train_scaled = X_scale.transform(X_train)
X_test_scaled = X_scale.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
model = Sequential()
```

```python
from tensorflow.keras.layers import Dense
#dense means every neuron conneted to every one in the previous layer
#relu is an activation fn
number_inputs = 40
number_hidden_nodes = 60
model.add(Dense(units=number_hidden_nodes,
                activation='relu', input_dim=number_inputs))
number_classes = 2
model.add(Dense(units=number_classes, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(
    X_train_scaled,
    one_hot_y_train,
    epochs=2000,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, one_hot_y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
