```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
exo = pd.read_csv('exoplanet_cleaned.csv')
exo
```

```python
#X and y values
#drop y vales and string object from df
X = exo.drop(columns=['koi_disposition']).values
y = exo["koi_disposition"].values

print(X.shape, y.shape)
X
```

```python
#separate 'training' and 'testing' data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
type(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=500)#default=100 did not work
classifier
```

```python
from sklearn.preprocessing import MinMaxScaler
X_scale = MinMaxScaler().fit(X_train)

X_train_scaled = X_scale.transform(X_train)
X_test_scaled = X_scale.transform(X_test)

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y = label_encoder.transform(y_train)
y_train, encoded_y
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
#penalty = ['l1', 'none']

params = {'penalty': ['l2', 'none']}
C = np.logspace(0,1)

grid = GridSearchCV(classifier, dict(C=C),cv=5, verbose=3)
grid
```

```python
#grid.fit(X_train_scaled, encoded_y)
#classifier.get_params().keys()
#grid.fit(X_train_scaled, encoded_y)
best_model = grid.fit(X_train_scaled, encoded_y)
```

```python
# View best hyperparameters
print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Best C:', best_model.best_estimator_.get_params()['C'])
# print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
# print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
print(best_model.best_params_)
print(best_model.best_score_)
```

```python
predictions = best_model.predict(X_test_scaled)
print(predictions)
```

```python
predictions = best_model.predict(X_test_scaled)
pd.DataFrame({"Prediction": label_encoder.inverse_transform(predictions), "Actual": y_test})
```

```python
from sklearn.metrics import classification_report
label_encoder.transform(y_test)
predictions
print(classification_report(label_encoder.transform(y_test), predictions))
```
